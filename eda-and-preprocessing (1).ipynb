# **Column Analysis**

### Life Expectency

We will check the life expectency column. 

Since many of our prections are based on life expectency we will drop the missing values in data. The number of missing rows in life_expectency column is 10.

```python
life_expect = life_expect.dropna(subset=['Life_Expectancy'])
```

```python
fig = px.histogram(life_expect, x="Life_Expectancy")
fig.show()
```

the maximum life expectency is 89 according to the dataset <br>
the minimum life expectency is 36.3 according to the dataset<br>
the standard deviation of life expectency is 9.52

```python
def numerical_column_analysis(col):
    print("The maximum {} is {} according to the dataset".format(col , life_expect[col].max()))
    print("The minimum {} is {} according to the dataset".format(col , life_expect[col].min()))
    print("the standard deviation of {} is {}".format(col , life_expect[col].std()))
    print("the median {} is {}".format(col , life_expect[col].median()))
    print("the mean of {} is {}".format(col , life_expect[col].mean()))
numerical_column_analysis("Life_Expectancy")
```

the mean life expectency is 69.22 while the median is 72.1 which means the column is left-skewed

### Adult Mortality and Infant Mortality

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(
    rows=2, cols=1,
    subplot_titles=("Adult_Mortality","Infant_Deaths"))
# fig = go.Figure(data=[go.Histogram(x=x)])
fig.append_trace(go.Histogram(x=life_expect["Adult_Mortality"]), row=1, col=1)
fig.append_trace(go.Histogram(x=life_expect["Infant_Deaths"]), row=2, col=1)
# fig.append_trace(go.Figure(life_expect=[go.Histogram(x="Infant_Deaths")]), row=2, col=1)

fig.show()
```

```python
print("Adult_morality")
numerical_column_analysis("Adult_Mortality")
print("Infant_Deaths")
numerical_column_analysis("Infant_Deaths")
```
